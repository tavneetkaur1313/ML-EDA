```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Price': [12000, 15000, 18000, 20000]
}

df = pd.DataFrame(data)

# Create a figure and axis
plt.figure(figsize=(12, 6))

# Create histograms
sns.histplot(df['Price'], color='blue', label='North', kde=True, bins=10)
sns.histplot(df['Price'], color='orange', label='South', kde=True, bins=10)
sns.histplot(df['Price'], color='green', label='East', kde=True, bins=10)
sns.histplot(df['Price'], color='red', label='West', kde=True, bins=10)

# Add titles and labels
plt.title('Distribution of Car Prices by Region')
plt.xlabel('Price (in $)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
